# **Github Actions**

- Github actions is a powerful automation tool provided by Github that enables developers to automate tasks related to thier software development workflows. It allows for the creation of custom workflows, which can be triggered by various events, such as pushesh to a repository, pull requests, or scheduled events. These workflows can be used for various purposes, including Continuos Integration (CI), Continuous Deployment (CD), testing and more.
- It is always provided by github
